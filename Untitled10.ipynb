```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generating sample data
np.random.seed(0)
years = np.arange(1, 11)
gdp = 1000 + 200 * years + np.random.normal(0, 150, 10)
unemployment_rate = 5 + 1.5 * years + np.random.normal(0, 1, 10)

# Create a DataFrame
data = pd.DataFrame({'Year': years, 'GDP': gdp, 'UnemploymentRate': unemployment_rate})
```

```python
# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
X = data['GDP'].values.reshape(-1, 1)
y = data['UnemploymentRate'].values
model.fit(X, y)

# Get the coefficients
slope = model.coef_[0]
intercept = model.intercept_

# Print the results
print(f"Slope (Coefficient): {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
```

```python
# Scatter plot
plt.scatter(data['GDP'], data['UnemploymentRate'], label='Data Points')

# Regression line
regression_line = model.predict(X)
plt.plot(data['GDP'], regression_line, color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('GDP')
plt.ylabel('Unemployment Rate')
plt.legend()

# Show the plot
plt.show()
```

```python
import torch
from transformers import PegasusForConditionalGeneration, PegasusTokenizer

# Load the Pegasus model and tokenizer
model_name = "google/pegasus-xsum"
model = PegasusForConditionalGeneration.from_pretrained(model_name)
tokenizer = PegasusTokenizer.from_pretrained(model_name)

# Input text
input_text = "Barcelona FC won the championship with a stunning victory over their rivals."

# Tokenize the input text
inputs = tokenizer(input_text, return_tensors="pt", max_length=1024, truncation=True)

# Generate the summary
summary_ids = model.generate(inputs["input_ids"], max_length=50, num_beams=4, length_penalty=2.0, early_stopping=True)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Print the generated summary
print("Generated Summary:")
print(summary)
```
